Noise is basically unpredicatable fluctuation 

## White Noise 
zero mean and constant standar deviation (correlation is 0 here) 

```python
import numpy as np 

# generate a time axis with sequential numbers upto 200
time = np.arange(200)
time 
```

```python
# samples
values = np.random.randn(200)*100 # normally distributed values mean 0 and std 1
values 
```

```python
import matplotlib.pyplot as plt 

plt.plot(time,values,label='Time Series Data')
plt.xlabel("Time (Units)")
plt.ylabel("Value")
plt.title("Example Time Series Plot")
plt.legend()
plt.show
```

## Red Noise 
it has zero mean and constant variance but is serially correlated in time unlike waves where precedding wave affects the successiding one 

```python
# setting the correlation coefficient 
r = 0.4
# generating the time axis 
time = np.arange(200)
# generating white noise 
white_noise = np.random.randn(200)*100
```

```python
# create red noise by introducing correlation between subsequent values in the white noise 
values = np.zeros(200)

for i,v in enumerate(white_noise):
    if i==0:
        values[i] = v   # first value will be the value of white noise only 
    else:
        values[i] = r*values[i-1] + np.sqrt((1-np.power(r,2))) * v  # then the values will update of values array that is initialised by 0 with respect to the index 
plt.plot(time,values,label = "Red Noise Process")
plt.legend()
plt.show()
```

x(j+1) = r * x(j) + np.sqrt(1-r*r) * w

here x (j+1) is the next value in the time series which is dependent on the current value     x(j) is the current value     

r is the correlation coefficient value (measure of strength -1 to 1)     

square root of (1 - r*r) * w adds some random variation to prevent the series from being a perfect copy of the past
