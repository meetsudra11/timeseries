```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating a time points (1 to 12)
t = np.arange(1, 13)
t
```

```python
# Generating data: Trend + Oscillations

trend = 10 + 2 * t  # y = c + mx 

# here x is each times step, m is the slope i.e data changes by 2 units at each time step and 10 is the intercept means starting value 
 
oscillation = 5 * np.cos((2 * np.pi / 4) * t)  # Periodic oscillation with period = 4
# The denominator (4) inside the cosine function tells us that the cycle repeats every 4 time units.

Y_t = trend + oscillation  # Total time series
```

```python
plt.figure(figsize=(3,3))
plt.plot(trend)
```

```python
plt.figure(figsize=(3,3))
plt.plot(t, oscillation, marker='s', linestyle='--', color='orange', label="Oscillation")
```

```python
plt.figure(figsize=(3,3))
plt.plot(Y_t)
```

```python

def moving_average(data, window):
    return np.convolve(data, np.ones(window)/window, mode='valid') 
#  np.convolve() is a general function that performs weighted sums


# Compute MAs
MA_4 = moving_average(Y_t, 4)  # Correct liya
MA_3 = moving_average(Y_t, 3)  # Incorrect liya
```

```python
plt.figure(figsize=(10,5))

plt.plot(t, Y_t, marker='o', label="Original Data (Trend + Oscillations)")
plt.plot(t[1:-1], MA_3, marker='s', linestyle="--", label="3-Point MA (Incorrect)")
plt.plot(t[1:-2], MA_4, marker='d', linestyle="--", label="4-Point MA (Correct)")

plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Effect of Choosing Correct vs Incorrect Moving Average Window")
plt.legend()
plt.grid()
plt.show()
```

Here we can see that the orange line with incorrect extent missed few oscillation and thus couldnt give correct trend
and the green line is perfectly linear excluding each oscillation
